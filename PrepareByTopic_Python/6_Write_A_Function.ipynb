```python
def is_leap(year):
    leap = False
    
    # Write your logic here
    if (year % 4 == 0 and year % 100 != 0) or year % 400 == 0:
        leap = True     
                
    return leap
    
year = int(input())
print(is_leap(year))
```

`note:`  
The second condition must be negated.  
The third condition can stand alone since anything divisible by 400 is also divisble by 4 and 100, making the third coniditon inclusive in the second and first condition.
